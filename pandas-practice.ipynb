## (2).Selection:

### (i) iloc(Integer Location):

```python
#DEFINITION:It is used to select rows and columns by their index number (position), not by name
```

```python
#df.iloc[0]-Select first row of data frame
print(df)
print(df.iloc[0])
```

```python
#df.iloc[1]-Select second row of data frame
df.iloc[1]
```

```python
#df.iloc[-1]-Select last row of data frame
df.iloc[-1]
```

```python
#df.iloc[:,0]-Select first column of data frame
print(df)
print(df.iloc[:,0])
```

```python
#df.iloc[:,1]-Select second column of data frame
df.iloc[:,1]  #Here 1 is a single integer not range or slicing, means it will exactly give the column at index 1
```

## (ii) loc:

```python
#DEFINITION: It is used to access rows and columns in a DataFrame by labels (names), not by numbers.
```

```python
import pandas as pd

data = {"Name": ["Amit", "Riya", "Sam"],
        "Age": [21, 22, 20]}
df = pd.DataFrame(data)
df
```

```python
#df.loc[rows,"column labels"]
print(df.loc[0,"Age"])  #Select single value by row position & column labels
```

```python
import pandas as pd

data = {"Name": ["Amit", "Riya", "Sam"],
        "Age": [21, 22, 20]}
df = pd.DataFrame(data,index=['row1','row2','row3'])
df
```

```python
#df.loc['row1':'row3', 'column1':'column3’]
#Both rows and columns are label-based (not index-based)

df.loc['row1':'row2','Name':'Age']   #Select and slicing on labels
```

# 4.Functions:


```python
import pandas as pd

dat3 = {"Math":[90,85,78,92,88],
        "Physics":[80,89,94,86,75],
        "Chemistry":[95,88,92,85,90]}
dat3 = pd.DataFrame(dat3)
dat3
```

## (i) Mean:

```python
dat3.mean()          #Mean of all columns
```

### (ii) Median:

```python
dat3.median()       #Median of each column
```

### (iii) Standard Deviation:

```python
dat3.std()        #Measures spread of data(How much values vary from the mean)
```

### (iv) Max:


```python
dat3.max()          #Largest value in each column
```

### (v) Min:

```python
dat3.min()            #Smallest value in each column
```

### (vi) Count:

```python
dat3.count()          #Number of non-null values in each column
```

### (vii) Describe:

```python
dat3.describe()      #Summary statistics for numerical columns
```
